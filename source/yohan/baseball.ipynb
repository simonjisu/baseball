```python
# Date 함수 사용


import urllib
import json
import datetime as dt
import time
```

```python
def date_count(interval):
    date_list = []
    for i in range(interval):
        yesterday = dt.datetime.now() - dt.timedelta(days=i+1)
        date_list.append(yesterday.strftime('%Y%m%d'))
        
    return date_list
```

```python
from sqlalchemy.ext.declarative import declarative_base

Base = declarative_base()

from sqlalchemy import Column, Integer, String
class User(Base):
    __tablename__ = 'users'

    id = Column(Integer, primary_key=True)
    name = Column(String)
    fullname = Column(String)
    password = Column(String)

    def __repr__(self):
        return "<User(name='%s', fullname='%s', password='%s')>" % (self.name, self.fullname, self.password)
```

```python
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker

yohanuser = User(name="Yohan", fullname="Yohan Park", password="1234")
asdfuser = User(name="asdf", fullname="asdf qwer", password="23445")

Session = sessionmaker()
Session.configure(bind=engine)  # once engine is available
session = Session()
session.add_all([yohanuser, asdfuser])
session.commit()
```

```python
from sqlalchemy.ext.declarative import declarative_base

Base = declarative_base()

from sqlalchemy import Column, Integer, String
class Livetext(Base):
    __tablename__ = 'livetext'

    id = Column(Integer, primary_key=True)
    batorder = Column(String)
    batter = Column(String)
    btop = Column(String)
    inning = Column(String)
    pitcher = Column(String)
    text = Column(String)
    textstyle = Column(String)

    def __repr__(self):
        return "Done"
```
